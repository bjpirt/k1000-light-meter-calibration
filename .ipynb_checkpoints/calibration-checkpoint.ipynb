# Pentax K1000 Light Meter Calibration

This notebook helps calibrate the built-in light meter of a Pentax K1000 by comparing its readings against a reference meter or known exposure values.

## Method
1. Take paired readings: K1000 meter EV vs reference EV (or known correct exposure)
2. Plot the relationship and fit a calibration curve
3. Determine the offset/correction factor to apply

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
````

## Helper Functions

EV (Exposure Value) is a log₂ scale. One stop = 1 EV.

To convert shutter speed + aperture to EV:
$$EV = \log_2\left(\frac{N^2}{t}\right)$$
where $N$ is f-number and $t$ is shutter speed in seconds.

````python
def aperture_shutter_to_ev(f_number, shutter_speed):
    """Convert aperture and shutter speed to EV.
    
    Args:
        f_number: aperture f-number (e.g. 5.6)
        shutter_speed: shutter speed in seconds (e.g. 1/125 = 0.008)
    """
    return np.log2(f_number**2 / shutter_speed)


def ev_to_stops_error(ev_measured, ev_reference):
    """Calculate error in stops between measured and reference EV."""
    return ev_measured - ev_reference


# Example: f/8 at 1/125s
print(f"f/8 at 1/125s = EV {aperture_shutter_to_ev(8, 1/125):.1f}")
print(f"f/5.6 at 1/60s = EV {aperture_shutter_to_ev(5.6, 1/60):.1f}")
````

## Enter Your Readings

Record paired readings from the K1000 meter and your reference.

For each reading, note the aperture and shutter speed the K1000 meter indicated at the needle-centred position, and the corresponding values from your reference meter (or a known correct exposure).

Edit the `readings` list below — each entry is:
```
(scene_description, k1000_f, k1000_shutter, reference_f, reference_shutter)
```
Or if you already have EV values, use the `ev_readings` list instead.

````python
# --- Enter your readings here ---
# Format: (description, k1000_f_number, k1000_shutter_s, reference_f_number, reference_shutter_s)
# shutter speed as a fraction e.g. 1/125 s = 1/125

readings = [
    # ("bright outdoor",  8,   1/250, 8,   1/500),
    # ("overcast outdoor", 5.6, 1/125, 5.6, 1/250),
    # ("indoor window",    2.8, 1/30,  2.8, 1/60),
]

# --- OR enter EV values directly ---
# Format: (description, k1000_ev, reference_ev)
ev_readings = [
    # ("bright outdoor", 14, 15),
    # ("overcast",       12, 13),
]

# Build dataframe from whichever list is populated
rows = []
for r in readings:
    desc, kf, ks, rf, rs = r
    rows.append({
        'scene': desc,
        'k1000_ev': aperture_shutter_to_ev(kf, ks),
        'reference_ev': aperture_shutter_to_ev(rf, rs),
    })
for r in ev_readings:
    desc, kev, rev = r
    rows.append({'scene': desc, 'k1000_ev': kev, 'reference_ev': rev})

df = pd.DataFrame(rows)
if not df.empty:
    df['error_stops'] = df['k1000_ev'] - df['reference_ev']
    display(df)
else:
    print("No readings entered yet. Add data to the 'readings' or 'ev_readings' lists above.")
````

## Analysis

````python
if df.empty:
    print("No data to analyse yet.")
else:
    # Linear regression: k1000_ev vs reference_ev
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        df['reference_ev'], df['k1000_ev']
    )
    
    mean_error = df['error_stops'].mean()
    std_error  = df['error_stops'].std()
    
    print("=== Calibration Summary ===")
    print(f"Number of readings : {len(df)}")
    print(f"Mean error         : {mean_error:+.2f} stops")
    print(f"Std deviation      : {std_error:.2f} stops")
    print(f"Linear fit slope   : {slope:.3f}  (ideal = 1.0)")
    print(f"Linear fit offset  : {intercept:+.3f} EV")
    print(f"R²                 : {r_value**2:.4f}")
    print()
    
    if abs(mean_error) < 0.25:
        print("✓ Meter reads within ±¼ stop on average — well calibrated.")
    elif abs(mean_error) < 0.5:
        print(f"⚠ Meter reads {abs(mean_error):.2f} stops {'high' if mean_error > 0 else 'low'} on average.")
        print(f"  Compensate by setting ISO {'+' if mean_error > 0 else '-'}{abs(mean_error):.1f} stops from box speed.")
    else:
        print(f"✗ Meter reads {abs(mean_error):.2f} stops {'high' if mean_error > 0 else 'low'} — significant calibration error.")
        print(f"  Compensate by setting ISO {'+' if mean_error > 0 else '-'}{abs(mean_error):.1f} stops from box speed,")
        print(f"  or consider servicing the meter.")
````

## Plots

````python
if df.empty:
    print("No data to plot yet.")
else:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # --- Plot 1: K1000 EV vs Reference EV ---
    ax = axes[0]
    ax.scatter(df['reference_ev'], df['k1000_ev'], s=80, zorder=5, label='Readings')
    
    ev_range = np.linspace(df['reference_ev'].min() - 1, df['reference_ev'].max() + 1, 100)
    ax.plot(ev_range, ev_range, 'k--', alpha=0.4, label='Perfect (1:1)')
    ax.plot(ev_range, slope * ev_range + intercept, 'r-', label=f'Fit (slope={slope:.2f})')
    
    for _, row in df.iterrows():
        ax.annotate(row['scene'], (row['reference_ev'], row['k1000_ev']),
                    textcoords='offset points', xytext=(5, 5), fontsize=8)
    
    ax.set_xlabel('Reference EV')
    ax.set_ylabel('K1000 Meter EV')
    ax.set_title('K1000 vs Reference Meter')
    ax.legend()
    ax.grid(True, alpha=0.3)

    # --- Plot 2: Error in stops ---
    ax2 = axes[1]
    colors = ['red' if e > 0 else 'steelblue' for e in df['error_stops']]
    bars = ax2.bar(df['scene'], df['error_stops'], color=colors, alpha=0.7)
    ax2.axhline(0, color='black', linewidth=0.8)
    ax2.axhline(mean_error, color='orange', linewidth=1.5, linestyle='--', label=f'Mean: {mean_error:+.2f} stops')
    ax2.axhspan(-0.5, 0.5, alpha=0.08, color='green', label='±½ stop')
    ax2.axhspan(-0.25, 0.25, alpha=0.12, color='green', label='±¼ stop')
    
    ax2.set_xlabel('Scene')
    ax2.set_ylabel('Error (stops)  [K1000 − Reference]')
    ax2.set_title('Meter Error by Scene')
    ax2.legend()
    ax2.grid(True, alpha=0.3, axis='y')
    plt.xticks(rotation=30, ha='right')

    plt.tight_layout()
    plt.savefig('calibration_results.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("Plot saved to calibration_results.png")
````

## ISO Compensation Table

If your meter reads consistently high or low, you can compensate by rating the film at a different ISO.
This table shows what ISO to set for common film speeds given the measured error.

````python
common_isos = [25, 50, 100, 200, 400, 800, 1600, 3200]

if df.empty:
    correction_stops = 0
    print("No data yet — showing table with 0 correction.")
else:
    correction_stops = -mean_error  # negate: meter reads high → rate film higher
    print(f"Meter error: {mean_error:+.2f} stops → ISO correction: {correction_stops:+.2f} stops\n")

rows_iso = []
for iso in common_isos:
    corrected = iso * (2 ** correction_stops)
    # Round to nearest standard ISO
    standard = min(common_isos, key=lambda x: abs(x - corrected))
    rows_iso.append({'Film ISO': iso, 'Set meter to': round(corrected, 1), 'Nearest standard': standard})

display(pd.DataFrame(rows_iso))
````
